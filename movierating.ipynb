```python
mean= df2.groupby('MovieId').agg({'Rating':'mean'})
```

```python
count=df2.groupby('MovieId').agg({'Rating':'count'})
```

```python
df1['count']= count
```

```python
df1['count']= df1['count'].round(0)
```

```python
df1.dropna(how='any', inplace=True)
```

```python
s= StandardScaler()
```

```python
df1.head()
```

```python
df1['rating']= np.where(df1['rating']>=3,1,0)
```

```python
df1['count']= s.fit_transform(df1[['count']])
```

```python
df1.drop(['MovieTitle'], axis=1, inplace=True)
```

```python
df1.drop(['ReleaseYear'], axis=1, inplace=True)
```

```python
df1.shape
```

```python
df1['MovieId']= s.fit_transform(df1[['MovieId']])
```

```python
y=df1['rating']
x= df1.drop(['rating'], axis=1)
```

```python
print(x)
```

```python
print(y)
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
lr= LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
train_pred= lr.predict(x_train)
```

```python
print(accuracy_score(y_train, train_pred))
```

```python
test_pred=lr.predict(x_test)
```

```python
print(accuracy_score(y_test, test_pred))
```
